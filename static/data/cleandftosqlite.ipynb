# Data SetUp

```python
#  Import Dependencies
import pandas as pd
import sqlite3
```

```python
raw_df=pd.read_csv("Resources/final_data.csv")
raw_df.head()
```

```python
encoded_df=pd.read_csv("Resources/encoded_features.csv").drop(["Unnamed: 0"], axis=1)
encoded_df.head()
```

## Migrating data from python to sql. 

```python
# Set up connection
conn=sqlite3.connect('Resources/db.sqlite')
```

```python
# Convert Pandas DataFrame to SQL Database
raw_df.to_sql('unitsSold_db', conn, if_exists='replace', index=False)
```

```python
#  Confirm data is in SQL database. 
pd.read_sql('select * from unitsSold_db', conn)
```

```python
# Convert Pandas DataFrame to SQL Database
encoded_df.to_sql('encodedUnitsSold_db', conn, if_exists='replace', index=False)
```

```python
#  Confirm data is in SQL database. 
pd.read_sql('select * from encodedUnitsSold_db', conn)
```
