# Data SetUp

```python
#  Import Dependencies
import pandas as pd
import sqlite3
```

```python
# Read in the CSV
df=pd.read_csv('Resources/Sales_Product_Price_by_Store.csv')
df=pd.DataFrame(df)
df
```

```python
# Creat a column to represent the 'Cost' of manufacturing, packaging, shipping the item.
df['Cost'] = df['Base Price']*.25

# Format Cost column to 2 decimal places
df['Cost'] = df['Cost'].map('{:.2f}'.format)

# Reorder the DF. 
df=df[['Store','Product','Date','Is_Holiday','Base Price','Price','Cost','Weekly_Units_Sold']]
df
```

```python
# Check data type of the Date column
df['Date'].dtype
```

```python
#  Separate Date column into Month and Day
df['Month'] = df['Date'].str.split('/').str[0].astype(int)
df['Day']=df['Date'].str.split('/').str[1].astype(int)
df
```

```python
# Date column is no longer needed, delete.
del df['Date']
df
```

```python
#  Reorder DF
df=df[['Store','Product','Month','Day','Is_Holiday','Base Price','Price','Cost','Weekly_Units_Sold']]
df
```

```python
# Same our working data to CSV format
df.to_csv("Resources/predictor.csv")
```

```python
#  Double check for missing or null values
df.isnull().sum()
```

## Migrating data from python to sql. 

```python
# Set up connection
conn=sqlite3.connect('db.sqlite')
```

```python
# Convert Pandas DataFrame to SQL Database
df.to_sql('profitPredictor_db', conn, if_exists='replace', index=False)
```

```python
#  Confirm data is in SQL database. 
pd.read_sql('select * from profitPredictor_db', conn)
```
